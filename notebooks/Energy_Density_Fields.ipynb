## Plots of the Energy Density Fields

#### Preamble

```python
import numpy as np
import astropy.io.fits as pyf
from astropy.io import fits
import matplotlib.pyplot as plt
```

#### Load in data

```python
ρ = fits.open('dens_t800.fits')[0].data
vx = fits.open('velx_t800.fits')[0].data
vy = fits.open('vely_t800.fits')[0].data
vz = fits.open('velz_t800.fits')[0].data
Bx = fits.open('magx_t800.fits')[0].data
By = fits.open('magy_t800.fits')[0].data
Bz = fits.open('magz_t800.fits')[0].data
```

#### Compute energy fields (kinetic, magnetic, and total)

```python
ρ = np.log(ρ)

v_dot_v = (vx**2 + vy**2 + vz**2)
v_dot_v = np.log(v_dot_v)

E_kinetic = ((ρ - np.mean(ρ)) / np.std(ρ)) * ((v_dot_v - np.mean(v_dot_v)) / np.std(v_dot_v))

B_dot_B = (Bx**2 + By**2 + Bz**2)
B_dot_B = np.log(B_dot_B)

E_B = (B_dot_B - np.mean(B_dot_B)) / np.std(B_dot_B)

E_total = E_B + E_kinetic
```

#### Make plots of kinetic energy, magnetic energy, and total energy

```python
plt.imshow(E_kinetic[:,:,0], cmap='Set3', vmin=-10, vmax=10)
plt.colorbar()
plt.title('Kinetic Energy')
plt.show()

plt.imshow(E_B[:,:,0], cmap='Set3', vmin=-10, vmax=10)
plt.colorbar()
plt.title('Magnetic Energy')
plt.show()

plt.imshow(E_total[:,:,0], cmap='gist_stern', vmin=-10, vmax=10)
plt.colorbar()
plt.title('Total Energy')
plt.show()
```
