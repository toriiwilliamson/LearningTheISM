# Plot of the Density Field of the Projected 4PCF

#### Preamble

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axes_grid1 as axgrid
from mpl_toolkits.axes_grid1 import make_axes_locatable
import astropy.io.fits as pyf
import os
import sarabande

plt.style.use("default")
```

#### Load in data (one time slice of a CATS MHD simulation, in this case t=800)

```python
hdulist = pyf.open('../data/dens_t800.fits.gz')
data = np.log(hdulist[0].data.astype(np.float64))
```

#### Reformat the data

```python
data = (data-np.mean(data))/(np.std(data))
```

#### Plot the projected density field

```python
fig, (ax1) = plt.subplots(1,1, figsize=(7,7))
im1 = ax1.imshow(data[:,:,64], cmap='hot', origin='lower')
divider1 = axgrid.make_axes_locatable(ax1)
cax1 = divider1.append_axes("right", size='5%', pad=0.05)
plt.colorbar(im1, cax=cax1)
ax1.set_title("Projected Density Field", fontsize=35, pad=15)
ax1.set_xlabel("x-axis", fontsize=35)
ax1.set_ylabel("y-axis", fontsize=35)
ax1.set_xticks([])
ax1.set_yticks([])
fig.tight_layout()
plt.show()
```
